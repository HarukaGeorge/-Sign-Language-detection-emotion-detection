## Data Prepration
    We used this module for applying gaussian blur to raw images and then split the data into train and test directories

## Gaussian Blur on Images

```python
# Importing Required Libraries

import numpy as np
import cv2
import os
from PIL import Image as im
import string

minValue = 70

# Function to convert an image to gaussian blur

def func(path):
    
    frame = cv2.imread(path)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray,(5,5),2)

    th3 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,2)
    ret, res = cv2.threshold(th3, minValue, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    print(type(ret),type(res))
    return res

# Setting initial path values (Source: dataset folder; Destination: final_dataset folder)

train_data_path = 'dataSet/trainingData'
target_path = 'final_dataSet/trainingData'
dir_0_path = 'dataSet/trainingData/'+str(0)
tgt_0_path = 'final_dataSet/trainingData/'+str(0)

# Gaussian BLur for blank images(folder: 0)

fig = ['{}.jpg'.format(i) for i in range(0,500)]

for fname in fig:
    src = os.path.join(dir_0_path, fname)
    result = func(src)
    g_fig = im.fromarray(result)
    tgt = os.path.join(tgt_0_path,fname)
    g_fig.save(tgt)

# Gaussian Blur for alphabets
    
for i in string.ascii_uppercase:
    src_inner_dir = os.path.join(train_data_path,i)
     
    tgt_inner_dir = os.path.join(target_path,i)
    
    print(src_inner_dir, tgt_inner_dir)
    fig = ['{}.jpg'.format(i) for i in range(0,1000)]
    
    for fname in fig:
        src = os.path.join(src_inner_dir, fname)
        result = func(src)
        g_fig = im.fromarray(result)
        tgt = os.path.join(tgt_inner_dir,fname)
        g_fig.save(tgt)
```

## Splitting images into train and test folder

```python
import glob
import shutil
import os
from string import ascii_uppercase

# Path to source(src_dir) and destination folder(dst_dir)

src_dir = "final_dataset/trainingData"
dst_dir = "final_dataset/testingData"

# Copying 250 images from source to detsination 

path_list = [os.path.join(src_dir, x) for x in ascii_uppercase]
path_list.append('final_dataset/trainingData\\0')
path_to_list = [os.path.join(dst_dir, x) for x in ascii_uppercase]
path_to_list.append('final_dataset/testingData\\0')
for n, jpgfile in enumerate(path_list):
    fnames = ['{}.jpg'.format(i) for i in range(1, 126)]
    for fname in fnames:
        src = os.path.join(jpgfile, fname)
        dst = os.path.join(path_to_list[n], fname)
        shutil.copyfile(src, dst)
    fnames = ['{}.jpg'.format(i) for i in range(500, 626)]
    for fname in fnames:
        src = os.path.join(jpgfile, fname)
        dst = os.path.join(path_to_list[n], fname)
        shutil.copyfile(src, dst)
        

# removing images from the train folder that are copied to test folder

for jpgfile in path_list:
    fnames = ['{}.jpg'.format(i) for i in range(1, 126)]
    for fname in fnames:
        src = os.path.join(jpgfile, fname)
        os.remove(src)
        
    fnames = ['{}.jpg'.format(i) for i in range(500, 626)]
    for fname in fnames:
        src = os.path.join(jpgfile, fname)
        os.remove(src)
    
```
